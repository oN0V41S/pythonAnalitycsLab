# Criando o Primeiro Gráfico e Regressão

```python
# Importando função de criar regressão
from sklearn.datasets import make_regression

# Criando regressão de x e y
# - n_samples -> número de amostras
# - noise -> ruído
x, y = make_regression(n_samples=200, n_features=1, noise=30)

# Importando função para gerar a regressão
import matplotlib.pyplot as plt

# Plotando dados da regressão
plt.scatter(x,y)
plt.show()
```
